```python
import numpy as np

def newton_step(x, eta):
    step = np.zeros(x.shape)
    for i in range(len(step)):
        step[i] = eta /(1/(1+x[i])**2 - 1/(1-x[i])**2 + 10e-4)
    return x - step

def check_eta_0(x_0, eta_0):
    step = np.zeros(x_0.shape)
    for i in range(len(step)):
        step[i] = eta_0  / (1 / (1 + x_0[i])**2 - 1 / (1 - x_0[i])**2 + 10e-4)
    norm = np.linalg.norm(step)
    return norm<=1/6

def IPM(x_0, eta_0, eps):
    m = len(x_0)
    T = int(np.ceil(np.log(m/(eps*eta_0))/np.log(1+1/(20*np.sqrt(m)))))
    x = x_0
    eta = eta_0
    for t in range(T):
        x = newton_step(x, eta)
        eta = eta*(1+1/(20*np.sqrt(m)))
    for i in range(2):
        x = newton_step(x, eta)
    return x

if __name__ == '__main__':
    x_0 = np.zeros((20,1))
    eta_0 = 1
    check_eta = check_eta_0(x_0, eta_0)
    while not check_eta:
        eta_0 /= 2
        check_eta = check_eta_0(x_0, eta_0)
    print(IPM(x_0, eta_0, 10e-6))

```
